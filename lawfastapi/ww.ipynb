행사 중 사고의 업무상 재해 인정 요건은 다음과 같습니다. 산업재해보상보험법 시행령 제30조에 따르면, 근로자가 행사에 참가하는 것이 사회통념상 노무관리 또는 사업운영상 필요하다고 인정되는 경우에 해당해야 합니다. 구체적으로는 다음과 같은 조건이 있습니다:\n\n### 1. 행사 참가의 필요성\n- 근로자가 행사에 참가하는 것이 사업주가 요구하거나, 사업의 운영상 필요하다고 인정되는 경우입니다. 예를 들어, 회사의 공식 행사나 교육, 훈련 등이 이에 해당할 수 있습니다.\n\n### 2. 사업주의 지시 또는 승인\n- 사업주가 근로자에게 행사에 참가하도록 지시했거나, 사전에 사업주의 승인을 받아 행사에 참가한 경우입니다. 이는 근로자가 자발적으로 참가한 것이 아니라, 사업주와의 관계에서 발생한 의무로 인식되어야 합니다.\n\n### 3. 행사 참가 중 발생한 사고\n- 근로자가 행사에 참가하는 과정에서 발생한 사고여야 하며, 이 사고가 업무와 관련이 있어야 합니다. 즉, 행사 중의 사고가 근로자의 업무 수행과 관련이 있어야 합니다.\n\n이러한 요건을 충족할 경우, 행사 중 발생한 사고는 업무상 재해로 인정될 수 있습니다. 따라서, 사고가 발생한 경위와 근로자의 행사 참가 여부, 사업주의 지시 등을 종합적으로 고려하여 판단하게 됩니다. 필요시 법률 전문가와 상담하여 구체적인 상황에 맞는 조언을 받는 것이 좋습니다.
